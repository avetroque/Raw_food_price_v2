```python
static = False
master_agg = []
for d in list_district:
    df_q = df_master.loc[df_master["district"]==d]
    list_item_by_district = df_q["item_code"].unique()
    #print(f'{d}:::{len(list_item_by_district)}')



    for item in list_item_by_district:
        x = df_q[df_q["item_code"]==item]
        max_ = x["price"].max()
        min_ = x["price"].min()
        mean_ = x["price"].mean()


        static = True if max_-min_== 0 else False
        #print(f'{item}, {max_}')
        master_agg.append([d,item,max_,min_,mean_,static])
        print(f'{d} {item} {max_} {min_} {mean_} {static}')
```

```python
master_agg_df = pd.DataFrame(master_agg,columns=['district','item_code','max_price','min_price','mean_price','static'])
master_agg_df.head()
```
