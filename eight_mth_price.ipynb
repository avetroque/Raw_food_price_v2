```python
URL_DATA = 'https://storage.googleapis.com/dosm-public-pricecatcher/lookup_premise.parquet'
df_premise_code = pd.read_parquet(URL_DATA)
if 'date' in df_premise_code.columns: df_premise_code['date'] = pd.to_datetime(df_premise_code['date'])

print(f'There are {df_premise_code["district"].nunique()} different districts')
print(f'There are {df_premise_code["state"].nunique()} different states')
#print(df_premise_code.head(10))



URL_DATA = 'https://storage.googleapis.com/dosm-public-pricecatcher/lookup_item.parquet'
df_item_code = pd.read_parquet(URL_DATA)
if 'date' in df_item_code.columns: df_item_code['date'] = pd.to_datetime(df_item_code['date'])

print(f'There are {df_item_code["item"].nunique()} different items')
print(f'There are {df_item_code["item_category"].nunique()} different item_categories')
#print(df_item_code.head())


item_category_list = df_item_code["item_category"].unique()
item_list = df_item_code["item"].unique()

print(item_category_list)

print(item_list)
```

```python

def map_item_premise_code(dataframe):
    tempdf = dataframe.copy() 
    
    s =tempdf.pop('premise_code').map(df_premise_code.set_index('premise_code')["state"])
    s.name = 'state'

    tempdf = dataframe.copy() 

    t = tempdf.pop('premise_code').map(df_premise_code.set_index('premise_code')['district'])
    t.name = "district"

    tempdf = dataframe.copy() 

    u = tempdf.pop('item_code').map(df_item_code.set_index('item_code')['item'])
    u.name = "item"

    tempdf = dataframe.copy() 

    v = tempdf.pop('item_code').map(df_item_code.set_index('item_code')['unit'])
    v.name = "unit"

    tempdf = dataframe.copy() 

    w = tempdf.pop('item_code').map(df_item_code.set_index('item_code')['item_category'])
    w.name = "item_category"

    df_master = pd.concat([dataframe,s,t,u,v,w],axis=1)
    df_master.sample(frac=1)

    return df_master

df1_master = map_item_premise_code(df1)
df2_master = map_item_premise_code(df2)
df3_master = map_item_premise_code(df3)
df4_master = map_item_premise_code(df4)
df5_master = map_item_premise_code(df5)
df6_master = map_item_premise_code(df6)
df7_master = map_item_premise_code(df7)
df8_master = map_item_premise_code(df8)
```

Let's focus on Minyak Masak Tulen Cap Alif 5kg (code 1942)

Seems that it only has data for 106 districts. Which minyak masak 5kg would cover all 163 districts?



```python
minyak5kg_list= [254, 257, 260, 263, 266, 1095, 1936, 1939, 1942]
    
def get_mean_price(dataframe_master,item_code):

    import math

    colLabel = ['District','Mean Price']
    mean_price = []

    i=1
    for district in district_list:
        newdf = dataframe_master.query('item_code==@item_code and district==@district')

        m_price = newdf["price"].mean()


        if math.isnan(m_price) == False: mean_price.append([district,m_price])
        
            
        i+=1
    
    pricedf = pd.DataFrame(mean_price,columns=colLabel)
    

    return pricedf

get_mean_price(df7_master,1942)

for item in minyak5kg_list: print(f' item_code:{item} mean_price:{len(get_mean_price(df4_master,item))}')    
for item in minyak5kg_list: print(f' item_code:{item} mean_price:{len(get_mean_price(df5_master,item))}')
for item in minyak5kg_list: print(f' item_code:{item} mean_price:{len(get_mean_price(df6_master,item))}')
for item in minyak5kg_list: print(f' item_code:{item} mean_price:{len(get_mean_price(df7_master,item))}')
```

```python
def get_range_price(dataframe_master,item_code):

    import math

    colLabel = ['District','Range Price']
    range_price = []

    i=1
    for district in district_list:
        newdf = dataframe_master.query('item_code==@item_code and district==@district')

        range = newdf["price"].max()-newdf["price"].min()


        if math.isnan(range) == False: range_price.append([district,range])
        
            
        i+=1
    
    rangedf = pd.DataFrame(range_price,columns=colLabel)
    

    return rangedf

range_df = get_range_price(df7_master,263)
```
